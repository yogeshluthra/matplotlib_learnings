```python
import matplotlib.pyplot as plt

n_rows=2
n_cols=2
plt.clf()

# This will split width and height of Figure into equal sizes. Note figure dimensions also play a role
# can be used to split unequally by adjusting the width_ratios and height_ratio arrays; and/or figure size
gridspecSQAURE={'width_ratios':[1]*max(n_rows,n_cols),
         'height_ratios':[1]*max(n_rows,n_cols)}    

F, grid = plt.subplots(n_rows, n_cols, gridspec_kw=gridspecSQAURE)#, sharex=True, sharey=True)
F.set_size_inches(8,8)  # 8 x 8 inches
F.suptitle('2x2 equal split', fontsize=16, x=0.5, y=0.98)

grid[0,0].plot([0, 1], [0, 1], 'k--')
grid[0,0].set_xlim([0.0, 1.0])
grid[0,0].set_ylim([0.0, 1.05])
grid[0,0].set_ylabel('class 0')
grid[0,0].set_title('ROC')
grid[0,0].set_aspect(1.0)
grid[0,1].set_title('PR')
grid[0,1].set_aspect(1.0)

plt.show()

#########
n_rows=4
n_cols=2
plt.clf()

# This will split width and height of Figure into equal sizes. Note figure dimensions also play a role
# can be used to split unequally by adjusting the width_ratios and height_ratio arrays; and/or figure size
gridspecSQAURE={'width_ratios':[1]*max(n_rows,n_cols),
         'height_ratios':[1]*max(n_rows,n_cols)}    

F, grid = plt.subplots(n_rows, n_cols, gridspec_kw=gridspecSQAURE)#, sharex=True, sharey=True)
F.set_size_inches(8,8)  # 8 x 8 inches
F.suptitle('4x4 equal split', fontsize=16, x=0.5, y=0.98)

grid[0,0].plot([0, 1], [0, 1], 'k--')
grid[0,0].set_xlim([0.0, 1.0])
grid[0,0].set_ylim([0.0, 1.05])
grid[0,0].set_ylabel('class 0')
grid[0,0].set_title('ROC')
grid[0,0].set_aspect(1.0)
grid[0,1].set_title('PR')
grid[0,1].set_aspect(1.0)

plt.show()

#########
n_rows=1
n_cols=2
plt.clf()

# This will split width and height of Figure into equal sizes. Note figure dimensions also play a role
# can be used to split unequally by adjusting the width_ratios and height_ratio arrays; and/or figure size
gridspecSQAURE={'width_ratios':[2,1],  # width of figure split 2 to 1
         'height_ratios':[1,1]}        # height of figure split equally in 2 chunks    

F, grid = plt.subplots(n_rows, n_cols, gridspec_kw=gridspecSQAURE)#, sharex=True, sharey=True)
F.set_size_inches(8,8)  # 8 x 8 inches
F.suptitle('1x2 unequal split', fontsize=16, x=0.5, y=0.98)

grid[0].plot([0, 1], [0, 1], 'k--')
grid[0].set_xlim([0.0, 1.0])
grid[0].set_ylim([0.0, 1.05])
grid[0].set_ylabel('class 0')
grid[0].set_title('ROC')
grid[0].set_aspect(1.0)
grid[0].set_title('PR')
grid[0].set_aspect(1.0)

plt.show()
```
